```python
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os
import pymongo
from pymongo.errors import DuplicateKeyError
```

```python
df = pd.read_csv(r"EasyVisa.csv")
df.head()
```

```python
df.shape
```

```python
for x in df.columns:
    print(x)
```

```python
data = df.to_dict(orient="records")
```

```python
len(data)
```

```python
print(data[0])
```

```python
load_dotenv()
username = os.getenv("MONGO_DB_USERNAME")
password = os.getenv("MONGO_DB_PASSWORD")
cluster = os.getenv("MONGO_CLUSTER_URL")

```

```python
DB_NAME = "US_VISA_PREDICTION"
COLLECTION_NAME = "VISA_APPROVAL_DATA"
CONNECTION_URL = (
    f"mongodb+srv://{username}:{password}@{cluster}/?appName=Cluster0"
)
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]



collection.create_index("case_id", unique=True)


try:
    collection.insert_many(data, ordered=False)
except DuplicateKeyError:
    print("Duplicate records skipped")



    
```
